### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(url, sep = '\t')
```

### Step 4. How many products cost more than $10.00?

```python
#Primero temos que observar que o tipo das informações contidas na coluna item_price é object
#Para trabalhar com as informações dessa coluna corretamente temos que convertê-la para o tipo float
chipo['item_price'].dtype
```

```python
#Usando uma função lamda para substituir o $ por um espaço vazio
chipo['item_price'] = chipo['item_price'].apply(lambda x: str(x).replace("$", " "))
```

```python
#Convertendo a coluna item_price para float
chipo['item_price'] = chipo['item_price'].astype('float64')
```

```python
#Observamos se a coluna foi convertida com sucesso
chipo['item_price'].dtype
```

```python
#Remover as repetições com o objetivo de contar apenas um único produto de cada
# A função .drop_duplicates - de acordo com a documentação do pandas - retornar o DataFrame removendo as linhas duplicadas
remove_repeticoes = chipo.drop_duplicates(['item_name','quantity'])
```

```python
#Utilizando dados com uma quantidade
#Nesses dados temos no item_price o preço de um único produto
produto_unico = remove_repeticoes[remove_repeticoes.quantity == 1]
```

```python
#Solicitando quantos produtos custam mais de $10.00
#A função .nunique() - de acordo com a documentação do pandas - conta observações distinhas sobre o eixo solicitado
produto_unico[produto_unico['item_price'] > 10].item_name.nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
#Removendo as repetições com o objetivo de trabalhar com apenas um único produto de cada
remove_repeticoes = chipo.drop_duplicates(['item_name','quantity'])
```

```python
#Usando dados com quantidade igual a 1.
#Nesses dados temos no item_price o preço de um único produto
produto_unico = remove_repeticoes[remove_repeticoes.quantity == 1]
```

```python
#Adicionando apenas as colunas item_name e item_price
price_of_each_item = produto_unico[['item_name', 'item_price']]
```

```python
#Solicitando uma tabela com a coluna item_name em ordem alfabética
price_of_each_item.sort_values(by = "item_name")
```

```python
#Solicitando uma tabela com a coluna item_price em ordem crescente
price_of_each_item.sort_values(by = "item_price")
```

### Step 6. Sort by the name of the item

```python
#Colocando a coluna item_name em ordem alfabética
chipo['item_name'].sort_values()
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(by = "item_price", ascending = False).head(1)
```
